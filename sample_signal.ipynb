```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parameters for the dataset
np.random.seed(42)
time_steps = np.linspace(0, 200, 2000)  # 1000 time points

# Create a composite signal
amplitude = (
    2 * np.sin(0.2 * time_steps)  # Slow oscillation
    + 0.5 * np.cos(0.8 * time_steps)  # Medium oscillation
    + 0.1 * np.sin(5 * time_steps)  # Fast oscillation (high-frequency noise)
    + np.random.normal(scale=0.2, size=time_steps.shape)  # Random noise
)

# Plot the dataset
plt.figure(figsize=(10, 5))
plt.plot(time_steps, amplitude, label="Synthetic Time Series")
plt.title("Complex Synthetic Time Series Data")
plt.xlabel("Time Steps")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()

# Save the dataset for further use
dataset = pd.DataFrame({'Time': time_steps, 'Amplitude':amplitude})
dataset.set_index('Time', inplace=True)
dataset.to_csv('sine_signal.csv')

```
